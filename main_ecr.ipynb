A continuación defino unas macros para automatizar la representación de los dataframes y figuras.

```julia
using CSV, DataFrames, DataFramesMeta, StatsPlots
using Plots.PlotMeasures
import Dates

macro head(in)
    quote
        display(@view $in[1:5, :])
    end
end

macro plot_style()
    quote
        xlabel!("Date")
        ylabel!("Temperature (°C)")
        plot!(legend=:outertop, gridlinewidth=3, framestyle=:box, dpi=500, left_margin=3mm)
    end
end
;
```

Se cargan los datos de temperatura en un DataFrame y se representan. Para ciertos tiempos no se tienen mediciones, lo que se refleja en los huecos del gráfico.

```julia
df = @chain CSV.read(
    "data/Temperatura de EMBALSE DE CHARCO REDONDO (CA) desde 2023-04-24 hasta 2023-04-30 - missingdata.csv",
    DataFrame;
    normalizenames=true,
    dateformat="yyyy-mm-dd HH:MM:SS",
) begin
    @select(:ts, :temp = :Temperatura_ºC_)
end

@head df

@df df plot(:ts, :temp, label="Input data", marker=:circle, markersize=2)
@plot_style
```

# Preparado de los datos

Antes de realizar la técnica de relleno de huecos, se decide "normalizar" los datos, restando la media y dividiendo por la desviación típica.

```julia
using Statistics

μ = mean(df[:, :temp])
σ = std(df[:, :temp])
@transform!(df, @byrow :temp_norm = (:temp - μ) / σ)

@df df plot(:ts, :temp_norm, marker=:circle, markersize=2, label="Input data")
@plot_style
ylabel!("Temperature, normalized")
```

Para el procesado de los datos, se necesita que para las zonas en las que no haya información, se tenga de forma explícita un `NaN`. Se crea un nuevo dataframe con filas de tiempo que rellenen homogéneamente desde todo el intervalo, y se le llama `df_resampled`. Después, se mapea para cada valor de tiempo, el valor de tiempo correspondiente del `df` original - si no se encuentra se coloca un `NaN`. En este caso me aprovecho de que los datos originales tambíen están equiespaciados igual que el rango que he construido. De no ser así, tendría que buscar la medida de tiempo más cercana, ver la diferencia de tiempos, etc; pero no hace falta en esta ocasión.  

Al dibujar el nuevo dataframe, Julia coloca huecos en los `NaN`, en vez de trazar una línea directa.

```julia
df_resampled = DataFrame(ts=range(
    df[:, :ts] |> minimum,
    df[:, :ts] |> maximum,
    step=Dates.Hour(1)
))

@transform!(df_resampled, @byrow :temp_norm = begin
    same_time = filter(row -> row.ts == :ts, df, view=true)

    if nrow(same_time) == 1
        return same_time.temp_norm[1]
    else
        return NaN
    end
end)


@head @subset(df_resampled, isnan.(:temp_norm))


@df df plot(:ts, :temp_norm, label="Input data")
@df df_resampled plot!(:ts, :temp_norm, linewidth=2, label="Resampled with NaN's")
@plot_style
ylabel!("Temperature, normalized")
```

# Predicción de la temperatura

Se utiliza la librería `StateSpaceModels`, en el que reimplementan una librería de R ([Automatic Time Series Forecasting: The forecast Package for R](https://doi.org/10.18637/jss.v027.i03)). Se usa un modelo ARIMA (Autoregressive Integrated Moving Average), es decir, una generalización de la combinación de los modelos autoregresivos y de media flotante. Por otra parte, esto es un modelo que necesita unos parámetros, por lo que se usa un calculador de parámetros del modelo ARIMA.



```julia
using StateSpaceModels

model = auto_arima(df_resampled.temp_norm)
fit!(model)
display(model)

smoother_output = kalman_smoother(model)

df_resampled[!, :temp_norm_predict] = get_smoothed_state(smoother_output)[:, 1]
@transform!(df_resampled, @byrow :temp = :temp_norm_predict * σ + μ)

plot()
@df df plot!(:ts, :temp, linewidth=1, label="Input data")
@df df_resampled plot!(:ts, :temp, linewidth=2, label=repr(model))
@plot_style
```

# Comparación

Por último, se compara el resultado de la predicción con los datos reales. Se abre el .csv con los datos originales y se dibujan uno encima de otro.

```julia
df_real = @chain CSV.read(
    "data/Temperatura de EMBALSE DE CHARCO REDONDO (CA) desde 2023-04-24 hasta 2023-04-30 - original.csv",
    DataFrame;
    normalizenames=true,
    dateformat="yyyy-mm-dd HH:MM:SS",
) begin
    @select(:ts, :temp = :Temperatura_ºC_)
end

@head df_real

plot()
@df df_real plot!(:ts, :temp, linewidth=2, label="Original data")
@df df_resampled plot!(:ts, :temp, linewidth=3, label=repr(model))
@plot_style
```

```julia
import StatsBase

df_both = innerjoin(
    @select(df_real, :ts, :temp_real = :temp), 
    @select(df_resampled, :ts, :temp_resampled = :temp, :temp_norm), 
    on=:ts
)

@head df_both
```

Se calculan el error medio absoluto (MAE) y la raíz del error cuadrático medio (RMSE) (en grados centígrados) para las secciones completas, y para exclusivamente las secciones reconstruidas.

```julia
errors_full = df_both |>
    (x -> full_range_errors = Dict(
        "mae" => StatsBase.meanad(x.temp_real, x.temp_resampled),
        "rmse" => StatsBase.rmsd(x.temp_real, x.temp_resampled),
    ))
```

```julia
errors_resample = df_both |>
    (x -> @subset(x, isnan.(:temp_norm))) |>
    (x -> full_range_errors = Dict(
        "mae" => StatsBase.meanad(x.temp_real, x.temp_resampled),
        "rmse" => StatsBase.rmsd(x.temp_real, x.temp_resampled),
    ))
```
